```python
# íŒŒì´ì¬ ì •ê·œí‘œí˜„ì‹ íŒ¨í‚¤ì§€ ì´ë¦„ì€ re ì…ë‹ˆë‹¤.
import re

# ì •ê·œì‹
# []: [] ì‚¬ì´ ë¬¸ìë¥¼ ë§¤ì¹˜, ^: not
regex = r"[^a-zA-Z0-9 ]"

# ì¹˜í™˜í•  ë¬¸ì
subst = ""

def tokenize(text):
    """text ë¬¸ìì—´ì„ ì˜ë¯¸ìˆëŠ” ë‹¨ì–´ ë‹¨ìœ„ë¡œ listì— ì €ì¥í•©ë‹ˆë‹¤.
    Args:
        text (str): í† í°í™” í•  ë¬¸ìì—´
    Returns:
        list: í† í°ì´ ì €ì¥ëœ ë¦¬ìŠ¤íŠ¸
    """
    # ì •ê·œì‹ ì ìš©
    tokens = re.sub(regex, subst, text)

    # ì†Œë¬¸ìë¡œ ì¹˜í™˜
    tokens = tokens.lower().split()
    
    return tokens
```

```python
def word_count(docs):
    """ í† í°í™”ëœ ë¬¸ì„œë“¤ì„ ì…ë ¥ë°›ì•„ í† í°ì„ ì¹´ìš´íŠ¸ í•˜ê³  ê´€ë ¨ëœ ì†ì„±ì„ ê°€ì§„ ë°ì´í„°í”„ë ˆì„ì„ ë¦¬í„´í•©ë‹ˆë‹¤.
    Args:
        docs (series or list): í† í°í™”ëœ ë¬¸ì„œê°€ ë“¤ì–´ìˆëŠ” list
    Returns:
        list: Dataframe
    """
    # ì „ì²´ ì½”í¼ìŠ¤ì—ì„œ ë‹¨ì–´ ë¹ˆë„ ì¹´ìš´íŠ¸
    word_counts = Counter()

    # ë‹¨ì–´ê°€ ì¡´ì¬í•˜ëŠ” ë¬¸ì„œì˜ ë¹ˆë„ ì¹´ìš´íŠ¸, ë‹¨ì–´ê°€ í•œ ë²ˆ ì´ìƒ ì¡´ì¬í•˜ë©´ +1
    word_in_docs = Counter()

    # ì „ì²´ ë¬¸ì„œì˜ ê°¯ìˆ˜
    total_docs = len(docs)

    for doc in docs:
        word_counts.update(doc)
        word_in_docs.update(set(doc))

    temp = zip(word_counts.keys(), word_counts.values())

    wc = pd.DataFrame(temp, columns = ['word', 'count'])

    # ë‹¨ì–´ì˜ ìˆœìœ„
    # method='first': ê°™ì€ ê°’ì˜ ê²½ìš° ë¨¼ì €ë‚˜ì˜¨ ìš”ì†Œë¥¼ ìš°ì„ 
    wc['rank'] = wc['count'].rank(method='first', ascending=False)
    total = wc['count'].sum()

    # ì½”í¼ìŠ¤ ë‚´ ë‹¨ì–´ì˜ ë¹„ìœ¨
    wc['percent'] = wc['count'].apply(lambda x: x / total)

    wc = wc.sort_values(by='rank')

    # ëˆ„ì  ë¹„ìœ¨
    # cumsum() : cumulative sum
    wc['cul_percent'] = wc['percent'].cumsum()

    temp2 = zip(word_in_docs.keys(), word_in_docs.values())
    ac = pd.DataFrame(temp2, columns=['word', 'word_in_docs'])
    wc = ac.merge(wc, on='word')
    
    # ì „ì²´ ë¬¸ì„œ ì¤‘ ì¡´ì¬í•˜ëŠ” ë¹„ìœ¨
    wc['word_in_docs_percent'] = wc['word_in_docs'].apply(lambda x: x / total_docs)

    return wc.sort_values(by='rank')
```

```python
def unlist(list):
    words=''
    for item in list:
        words+=item+' '
    return words
```

```python
def compute_vader_scores(df, label):
    sid = SentimentIntensityAnalyzer()
    df["vader_neg"] = df[label].apply(lambda x: sid.polarity_scores(unlist(x))["neg"])
    df["vader_neu"] = df[label].apply(lambda x: sid.polarity_scores(unlist(x))["neu"])
    df["vader_pos"] = df[label].apply(lambda x: sid.polarity_scores(unlist(x))["pos"])
    df["vader_comp"] = df[label].apply(lambda x: sid.polarity_scores(unlist(x))["compound"])
    df['cleantext2'] = df[label].apply(lambda x: unlist(x))
    return df
```

```python
def tweet_to_words(tweet):
    ''' Convert tweet text into a sequence of words '''
    
    # convert to lowercase
    text = tweet.lower()
    # remove non letters
    text = re.sub(r"[^a-zA-Z0-9]", " ", text)
    # tokenize
    words = text.split()
    # remove stopwords
    words = [w for w in words if w not in stopwords.words("english")]
    # apply stemming
    words = [PorterStemmer().stem(w) for w in words]
    # return list
    return words
```

```python
import tensorflow.keras.backend as K

def f1_score(precision, recall):
    ''' Function to calculate f1 score '''
    
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val
```
